##### Generating "no fire" data  
For a classification prediction problem, we need training data that are labeled as "fire" and "no fire"

```python
import bisect
from datetime import timedelta

# Time range
start_time = pd.to_datetime("2024-01-01")
end_time = pd.to_datetime("2024-12-31")
total_seconds = int((end_time - start_time).total_seconds())

def generate_random_times(n, existing_times):
    existing_times = sorted(existing_times)
    generated = []

    # Convert to list of timestamps in seconds since epoch for fast math
    existing_seconds = [pd.to_datetime(t).timestamp() for t in existing_times]
    generated_seconds = []

    attempts = 0
    max_attempts = 5000
    min_separation = 86400  # 24 hours in seconds

    while len(generated_seconds) < n and attempts < max_attempts:
        candidate = start_time + timedelta(seconds=np.random.randint(0, total_seconds))
        candidate_sec = candidate.timestamp()

        # Use bisect to insert into sorted list
        all_times = sorted(existing_seconds + generated_seconds)
        idx = bisect.bisect_left(all_times, candidate_sec)

        too_close = False
        if idx > 0 and abs(candidate_sec - all_times[idx - 1]) < min_separation:
            too_close = True
        elif idx < len(all_times) and abs(candidate_sec - all_times[idx % len(all_times)]) < min_separation:
            too_close = True

        if not too_close:
            generated.append(candidate)
            generated_seconds.append(candidate_sec)

        attempts += 1

    if len(generated_seconds) < n:
        print(f"Warning: Only generated {len(generated_seconds)} out of {n} due to 24-hour proximity rule.")

    return generated
```

```python
new_rows = []
loc_time_df = pd.read_csv("wildfire_location_time_fuel.csv")
loc_time_df['LATITUDE_round'] = loc_time_df['LATITUDE'].round(1)
loc_time_df['LONGITUDE_round'] = loc_time_df['LONGITUDE'].round(1)
for (lat, lon), time_df in loc_time_df.groupby(['LATITUDE_round', 'LONGITUDE_round']):
    count = len(time_df)
    if count >= 10:
        num_to_generate = 1
    else:
        num_to_generate = 10 - count

    new_times = generate_random_times(num_to_generate, time_df['FIRE_START_DATE'].tolist())
    for t in new_times:
        # append latitude and longitude of the original entry
        new_rows.append({
            'LATITUDE': time_df['LATITUDE'].values[0],
            'LONGITUDE': time_df['LONGITUDE'].values[0],
            'FIRE_START_DATE': t,
            'FUEL_TYPE': time_df['FUEL_TYPE'].values[0]
        })

# Create DataFrame of new entries
df_new = pd.DataFrame(new_rows)
df_new.to_csv("no_fires_augmented_fuel1.csv", index=False)

# Combine with original data
df_augmented = pd.concat([loc_time_df[['LATITUDE', 'LONGITUDE', 'FIRE_START_DATE', 'FUEL_TYPE']], df_new], ignore_index=True)
display(df_augmented) # 26551 rows + new rows
df_augmented.to_csv("fires_no_fires_augmented_fuel1.csv", index=False)
```

```python
# DEPRICATED!!! DO NOT USE THIS FUNCTION!!! DO NOT INCLUDE THIS IN FINAL REPORT!!!
def add_lightning_info_to_weather(weather_df, lightning_df,
                                  lightning_lat_col="CENTERLAT", lightning_lon_col="CENTERLON",
                                  lightning_date_col="ZDAY",
                                  weather_lat_col="LATITUDE", weather_lon_col="LONGITUDE",
                                  weather_date_col="START_DATE"):
    # Work on copies to avoid modifying originals
    weather_df = weather_df.copy()
    lightning_df = lightning_df.copy()

    # Parse dates in both DataFrames
    lightning_df["Date"] = pd.to_datetime(lightning_df[lightning_date_col], format="%Y%m%d").dt.date
    weather_df["ParsedDate"] = pd.to_datetime(weather_df[weather_date_col], format="%Y-%m-%d %H:%M:%S", errors="coerce").dt.date
    weather_df = weather_df[weather_df["ParsedDate"].notna()]

    ## --- 0.1Â° Match (Exact with rounding) ---
    # Create rounded coordinates for lightning data
    lightning_df["lat_r"] = lightning_df[lightning_lat_col].round(1)
    lightning_df["lon_r"] = lightning_df[lightning_lon_col].round(1)
    # Aggregate TOTAL_COUNT by Date and rounded coordinates
    lightning_agg = (lightning_df
                     .groupby(["Date", "lat_r", "lon_r"])["TOTAL_COUNT"]
                     .sum()
                     .reset_index())

    # Create rounded coordinates for weather data
    weather_df["lat_r"] = weather_df[weather_lat_col].round(1)
    weather_df["lon_r"] = weather_df[weather_lon_col].round(1)

    # Merge the aggregated lightning counts with weather data on date and rounded lat/lon
    weather_df = weather_df.merge(lightning_agg,
                                  left_on=["ParsedDate", "lat_r", "lon_r"],
                                  right_on=["Date", "lat_r", "lon_r"],
                                  how="left")
    # Fill missing counts with 0
    weather_df["Lightning_Count_0.1deg"] = weather_df["TOTAL_COUNT"].fillna(0)

    # # Optionally, drop intermediate columns
    weather_df.drop(columns=["Date", "lat_r", "lon_r"], inplace=True, errors="ignore")

    return weather_df
```

population density of nearest urban center and distance to it

```python
cities_df = pd.read_csv("./PopulationData/canadian_cities_with_coords.csv")
display(cities_df) # 119 rows

from scipy.spatial import cKDTree

urban_coords = cities_df[['latitude', 'longitude']].to_numpy()
urban_tree = cKDTree(urban_coords)

# Compute nearest urban center
def get_distance_to_urban(lat, lon):
    dist, idx = urban_tree.query([lat, lon], k=1)
    # also return the population density, dwelling count, and urban city type of the nearest urban center using its idx
    urban_density = cities_df.iloc[idx]['Population density per square kilometre, 2021']
    dwellings = cities_df.iloc[idx]['Total private dwellings, 2021']
    urban_type = cities_df.iloc[idx]['Geographic area type']
    # Geographic area types include 'Small population centre', 'Medium population centre', and 'Large urban population centre'
    # these labels are independent of the population density, for example the population density in Calgary (Large) is smaller than the population density in Airdrie (medium) and Chestermere (small)

    return dist, urban_type, urban_density, dwellings  # in degrees, multiply by ~111 km/deg for rough estimate

# Apply to your df
wildfire_df[['dist_to_urban_deg', 'urban_type', 'urban_population_density', 'urban_dwelling']] = pd.DataFrame(
    wildfire_df.apply(lambda row: get_distance_to_urban(row['LATITUDE'], row['LONGITUDE']), axis=1).tolist(),
    index=wildfire_df.index
)
wildfire_df['dist_to_urban_km'] = wildfire_df['dist_to_urban_deg'] * 111  # rough km conversion

display(wildfire_df)
```
